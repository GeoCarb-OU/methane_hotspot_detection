```python
from tqdm import trange
import numpy as np
from random import randint
hotspot_img = []
non_hp_img = []

for idx in trange(len(XC_normal)):
    emissions = np.array(EM_normal[idx])
    surface_p = np.array(SP_normal[idx])
    dry_air = np.array(XC_normal[idx])
    em_data = emissions[:,:] > 0.005 #Threshold
    xs, ys = em_data.nonzero()
    em_data = np.invert(em_data)
    nxs,nys = em_data.nonzero()
    for i in range(len(xs)):
        j = randint(20, len(nxs) - 20)
        hotspot_img.append(dry_air[(xs[i] - 10):(xs[i] +10), (ys[i] -10) : (ys[i] +10)])
        part = dry_air[(nxs[j] - 10):(nxs[j] +10), (nys[j] -10) : (nys[j] +10)]
        
        while(1):
            if part.shape == (20,20):
                       non_hp_img.append(part)
                       break
            else:
                       j = randint(20, len(nxs) - 20)
                       part = dry_air[(nxs[j] - 10):(nxs[j] +10), (nys[j] -10) : (nys[j] +10)]
```

```python
# Save data

import pickle 
with open("Hotspot.pkl", "wb") as fp:
  pickle.dump(hotspot_img, fp)
with open("Non_hotspot.pkl", "wb") as fn:
  pickle.dump(non_hp_img,fn)
```
